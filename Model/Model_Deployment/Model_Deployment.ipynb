# Create Flask Application

```python
import numpy as np
from flask import Flask, request, render_template
import pickle

app = Flask(__name__)
run_with_ngrok(app)
model = pickle.load(open('model_ITS04_09.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/getprediction',methods=['POST'])
def getprediction():    

    input = [int(x) for x in request.form.values()]
    final_input = np.array(input)
    final_input = final_input.reshape(1, -1)
    #print(final_input)
    #print(type(final_input))
    #print(final_input.shape)
    prediction = model.predict(final_input)

    return render_template('index.html', output='Predicted Customers Cluster is Cluster {}'.format(prediction))
   

app.run()
```
